```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
np.set_printoptions(suppress = True,precision = 9)


x_values = np.round(np.arange(0.001, 0.900, 0.020), 2)
y_values = np.round(np.arange(0.001, 0.900, 0.020), 2)
fig, ax = plt.subplots(figsize = (8, 6))
heatmap = ax.imshow(np.load("results.npy")[0][0], cmap = 'magma', origin = 'upper', aspect = 'auto')
# Set major tick positions only at every 5th index
x_tick_positions = np.arange(0, len(x_values), 5)  # Indices at a gap of 5
y_tick_positions = np.arange(0, len(y_values), 5)

ax.set_xticks(x_tick_positions)
ax.set_yticks(y_tick_positions)

ax.set_xticklabels(x_values[x_tick_positions])
ax.set_yticklabels(y_values[y_tick_positions])

# Hide minor ticks (remove small markers between labeled ticks)
ax.xaxis.set_minor_locator(plt.NullLocator())
ax.yaxis.set_minor_locator(plt.NullLocator())
# Add colorbar
plt.colorbar(heatmap)
# Label the axes
ax.set_xlabel(r"$\lambda_2$" + "\n (f)", fontsize = 15)
ax.set_ylabel(r"$\lambda_1$", fontsize = 15)
ax.set_title("Relative reconstruction error", fontsize = 15)
plt.grid()
# plt.show()
plt.savefig("HyperparameterSensitivity.pdf")
```
